```python
import tkinter as tk
from tkinter import ttk, messagebox
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from imblearn.over_sampling import SMOTE

# Preprocess data function (as in your original code)
def preprocess_data(data):
    label_encoder = LabelEncoder()
    data['gender'] = label_encoder.fit_transform(data['gender'])
    data['ever_married'] = label_encoder.fit_transform(data['ever_married'])
    data['work_type'] = label_encoder.fit_transform(data['work_type'])
    data['Residence_type'] = label_encoder.fit_transform(data['Residence_type'])
    data['smoking_status'] = label_encoder.fit_transform(data['smoking_status'])
    data = data.fillna(data.mean())
    return data

# Balance dataset function (as in your original code)
def balance_dataset(X, y):
    sm = SMOTE(random_state=42)
    X_res, y_res = sm.fit_resample(X, y)
    return X_res, y_res

# Train models functions (as in your original code)
def train_logistic_regression(X_train, y_train):
    lr = LogisticRegression(max_iter=1000)
    lr.fit(X_train, y_train)
    return lr

def train_random_forest(X_train, y_train):
    classifier_rf = RandomForestClassifier(n_estimators=100, random_state=42)
    classifier_rf.fit(X_train, y_train)
    return classifier_rf

# Make predictions function
def make_predictions(model, input_data):
    input_data = preprocess_data(input_data)
    scaler = MinMaxScaler()
    X_res = scaler.fit_transform(input_data)
    return model.predict(X_res)

# Button click event function
def on_predict():
    try:
        # Get input values
        input_data = {
            'gender': [int(gender_var.get())],
            'age': [float(age_var.get())],
            'hypertension': [int(hypertension_var.get())],
            'heart_disease': [int(heart_disease_var.get())],
            'ever_married': [int(ever_married_var.get())],
            'work_type': [work_type_var.get()],
            'Residence_type': [int(Residence_type_var.get())],
            'avg_glucose_level': [float(avg_glucose_level_var.get())],
            'bmi': [float(bmi_var.get())],
            'smoking_status': [int(smoking_status_var.get())]
        }
        input_df = pd.DataFrame(input_data)

        # Make predictions
        if model_choice.get() == 'Logistic Regression':
            predictions = make_predictions(logistic_regression_model, input_df)
        elif model_choice.get() == 'Random Forest':
            predictions = make_predictions(random_forest_model, input_df)
        
        messagebox.showinfo("Prediction Result", f"Stroke Prediction: {'Yes' if predictions[0] == 1 else 'No'}")
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numerical values.")

# Load and preprocess dataset
data = pd.read_csv('healthcare-dataset-stroke-data.csv')
data = preprocess_data(data)
X = data.drop(['stroke'], axis=1)
y = data['stroke']
X_res, y_res = balance_dataset(X, y)

# Train models
logistic_regression_model = train_logistic_regression(X_res, y_res)
random_forest_model = train_random_forest(X_res, y_res)

# GUI setup
window = tk.Tk()
window.title("Stroke Prediction")
window.geometry("400x500")

# Input labels and fields
labels = [
    "Gender (Male=0, Female=1)", "Age", "Hypertension (0=No, 1=Yes)", 
    "Heart Disease (0=No, 1=Yes)", "Ever Married (0=No, 1=Yes)", 
    "Work Type (Choose one)", "Residence Type (0=Urban, 1=Rural)",
    "Average Glucose Level", "BMI", 
    "Smoking Status (0=Unknown, 1=Never, 2=Formerly, 3=Smokes)"
]
variables = [tk.StringVar() for _ in range(len(labels))]
entries = []

for i, label in enumerate(labels):
    tk.Label(window, text=label).grid(row=i, column=0, sticky="w", padx=10, pady=5)
    if i == 5:
        # Dropdown for Work Type
        work_type_choices = ['Private', 'Self-employed', 'Govt_job', 'children', 'Never_worked']
        work_type_var = ttk.Combobox(window, values=work_type_choices, textvariable=variables[i])
        work_type_var.grid(row=i, column=1, padx=10, pady=5)
        entries.append(work_type_var)
    else:
        entry = tk.Entry(window, textvariable=variables[i])
        entry.grid(row=i, column=1, padx=10, pady=5)
        entries.append(entry)

# Model selection dropdown
tk.Label(window, text="Select Model:").grid(row=len(labels), column=0, padx=10, pady=10)
model_choice = ttk.Combobox(window, values=['Logistic Regression', 'Random Forest'])
model_choice.grid(row=len(labels), column=1, padx=10, pady=10)
model_choice.current(0)

# Predict button
predict_button = tk.Button(window, text="Predict", command=on_predict)
predict_button.grid(row=len(labels)+1, column=1, pady=20)

window.mainloop()
```
