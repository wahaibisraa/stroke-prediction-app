```python
from flask import Flask, request, jsonify
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pickle

app = Flask(__name__)

# Dummy data preparation and model training for demonstration
data = pd.DataFrame({
    'age': [65, 45, 25],
    'hypertension': [1, 0, 1],
    'heart_disease': [0, 1, 0],
    'avg_glucose_level': [120.5, 80.2, 90.5],
    'bmi': [35.0, 25.0, np.nan]
})
data['bmi'] = data['bmi'].fillna(data['bmi'].mean())
features = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']

# Scale features and fit model
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[features])
model = RandomForestClassifier()
model.fit(data_scaled, [0, 1, 0])

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()

    try:
        # Extract and convert input data
        input_data = [
            float(data['age']),
            int(data['hypertension']),
            int(data['heart_disease']),
            float(data['avg_glucose_level']),
            float(data['bmi'])
        ]
        # Scale input data
        input_data_scaled = scaler.transform([input_data])

        # Predict
        prediction = model.predict(input_data_scaled)
        result = "High risk of stroke" if prediction[0] == 1 else "Low risk of stroke"

        return jsonify({"prediction": result})
    except Exception as e:
        return jsonify({"error": str(e)}), 400

if __name__ == "__main__":
    app.run(debug=True)
```
